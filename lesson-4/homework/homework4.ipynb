Dictionary and Set exercises

```python
#1
ages = {
    "Alice": 25,
    "Bob": 30,
    "Charlie": 35 
}

x = ages.values()
x = sorted(x, reverse=True) # descending order
y = sorted(x) # ascending order
print(x,y)
```

```python
#2
sample_dict = {
    0 : 10,
    1 : 20
}

sample_dict[2] = 30

print(sample_dict)
```

```python
#3
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

dic4 = dic1 | dic2 | dic3
print(dic4)
```

```python
#4
new_dict = {}
n = int(input('Son kiriting:'))

for i in range(1,n+1):
    new_dict[i] = i**2
print(new_dict)

```

```python
#5
new_dict = {}

for i in range(15):
    new_dict[i+1] = (i+1)**2
print(new_dict)
```
